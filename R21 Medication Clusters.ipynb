```python
# Process data and extract the existing medication IDs in the ML library for different patient_DEID in the med_admin file

med_list_v3 = med_list_v2.dropna(subset = ["UNC"])
med_list_v3 = med_list_v3[med_list_v2.UNC != "Unavailable"]
med_list_v3 = med_list_v3.reset_index()
med_list_v3 = med_list_v3.drop(['index'], axis=1)
 
unique_meds = med_list_v3['UNC'].unique()

med_admin_v3 = med_admin_v2[med_admin_v2["MEDICATION_ID"].isin(unique_meds)][["PATIENT_DEID","MEDICATION_ID"]]
```

```python
# Select additional features form the master table file

selected_feature = master_table_v2[["PATIENT_DEID","race_comb", "age_at_time", "SEX", "total_length_days",
                                    "Death_at_discharge", "score_24h","length_of_ICUstay_days", "score_lastday",
                                    "Tertile_24h", "MV_in24h_fac","Mechanical_ventilation", 
                                    "fluid_overload_wt", "fluid_overload_overall","fluid_overload_volume",
                                    "Delirium_ICU","AKI","admission_dx_type", "main_ICU_type", "APACHE_24h_score", 
                                    "vasopressor_ICU_day_0_included","cum_fluid_day1","CRRT","Weight_kg"]]

selected_feature_v2 = selected_feature.copy()
#selected_feature_v2['race_comb'].replace(['White or Caucasian','Black or African American', 'Other'],[1, 2, 3], inplace=True)
#selected_feature_v2['SEX'].replace(['Female','Male'],[1, 2], inplace=True)
selected_feature_v2['Death_at_discharge'].replace(['dead_at_discharge','alive'],[0, 1], inplace=True)
selected_feature_v2['Tertile_24h'].replace(['low','middle', 'high'],[0, 1, 2], inplace=True)
selected_feature_v2['MV_in24h_fac'].replace(['none','before 24h', 'after 24h'],[0, 1, 2], inplace=True)
selected_feature_v2['Mechanical_ventilation'].replace(['No','Yes'],[0, 1], inplace=True)
selected_feature_v2['fluid_overload_overall'].replace(['No','Yes'],[0, 1], inplace=True)
selected_feature_v2['Delirium_ICU'].replace(['Negative','Positive'],[0, 1], inplace=True)
selected_feature_v2['AKI'].replace(['No','Yes'],[0, 1], inplace=True)
selected_feature_v2['main_ICU_type'].replace(['Medical','Cardiac','Cardiothoracic Surgery', 'Surgical', 'Burn','Neurosciences','Mixed'],[1, 2, 2, 3, 4, 5, 6], inplace=True)
selected_feature_v2['CRRT'].replace(['No','Yes'],[0, 1], inplace=True)

#selected_feature.loc[:,"dur_MV_days"] = selected_feature["dur_MV_days"].fillna(0)
#selected_feature.loc[:,"vasopressor_ICU_day"] = selected_feature["vasopressor_ICU_day"].fillna(0)
#selected_feature.total_length_days = selected_feature.total_length_days.fillna(0)
#selected_feature.first_MV = selected_feature.first_MV.fillna(0)

#selected_feature = selected_feature.fillna("Unknown")
```

```python
selected_feature_v3 = selected_feature_v2.copy()


selected_feature_v3.loc[:,"MV_in24h_fac"] = selected_feature_v3["MV_in24h_fac"].fillna(0)
selected_feature_v3.loc[:,"Mechanical_ventilation"] = selected_feature_v3["Mechanical_ventilation"].fillna(0)
selected_feature_v3.loc[:,"fluid_overload_wt"] = selected_feature_v3["fluid_overload_wt"].fillna(0)
selected_feature_v3.loc[:,"fluid_overload_overall"] = selected_feature_v3["fluid_overload_overall"].fillna(0)
selected_feature_v3.loc[:,"fluid_overload_volume"] = selected_feature_v3["fluid_overload_volume"].fillna(0)
selected_feature_v3.loc[:,"Delirium_ICU"] = selected_feature_v3["Delirium_ICU"].fillna(0)
selected_feature_v3.loc[:,"AKI"] = selected_feature_v3["AKI"].fillna(0)
selected_feature_v3.loc[:,"vasopressor_ICU_day_0_included"] = selected_feature_v3["vasopressor_ICU_day_0_included"].fillna(0)
selected_feature_v3.loc[:,"CRRT"] = selected_feature_v3["vasopressor_ICU_day_0_included"].fillna(0)
```

```python
selected_feature_v4 = selected_feature_v3.copy()
selected_feature_v4.set_index('PATIENT_DEID', inplace=True, drop=True)
selected_feature_v4
```

```python
select = selected_feature_v4.to_csv("select.csv", index = False)
```

```python
import miceforest as mf
# MICE using LGB 
kds = mf.ImputationKernel(
  selected_feature_v4,
  save_all_iterations=True,
  random_state=3
)

# Run the MICE algorithm for 2 iterations
kds.mice(5)

# Return the completed dataset.
selected_feature_v4 = kds.complete_data()
```

```python
# Combine selected features with medication IDs and patient_DEID

IDs = med_admin_v3[["PATIENT_DEID","MEDICATION_ID"]]
feature_combined = pd.merge(IDs, selected_feature_v4, on ="PATIENT_DEID")
```

```python
feature_combined = feature_combined.astype({"MEDICATION_ID":'int'}) 
IDs_v3 = IDs_v2.astype({"MEDICATION_ID":'int'}) 
IDs_v4 = IDs_v3.drop_duplicates()
IDs_v4
```

```python
meds_per_id = IDs_v4.groupby("PATIENT_DEID")["MEDICATION_ID"].apply(list).reset_index(name = "GIVEN_MEDS") #just unique meds
meds_per_id
```

```python
meds_per_id.iloc[0,1]
```

```python
meds_per_id_v2 = meds_per_id.copy()
master_patients = np.array(selected_feature_v3["PATIENT_DEID"])
meds_per_id_v3 = meds_per_id_v2[meds_per_id_v2["PATIENT_DEID"].isin(master_patients)][["PATIENT_DEID","GIVEN_MEDS"]]
```

```python
meds_per_id_v3 = meds_per_id_v3.reset_index()
meds_per_id_v3
```

```python
#meds_per_id_v2.set_index('PATIENT_DEID', inplace=True, drop=True)
# Create binary set
from mlxtend.preprocessing import TransactionEncoder

# PROJECT EACH RECORD'S DX CODES
aki_meds = meds_per_id_v3.GIVEN_MEDS.to_list()
te = TransactionEncoder()
te_ary = te.fit(aki_meds).transform(aki_meds)
final_df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
final_df_v2 = final_df.copy()
final_df_v2 = pd.concat([meds_per_id_v3['PATIENT_DEID'], final_df_v2], axis = 1)
final_df_v2.set_index(final_df_v2['PATIENT_DEID'], inplace=True, drop=True)
final_df_v2 = final_df_v2.iloc[: , 2:]*1
final_df_v2
```

```python
final_data = pd.merge(selected_feature_v4, final_df_v2, on ="PATIENT_DEID")
```
